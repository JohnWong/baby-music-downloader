```python
import urllib.request
import json
import os
import shutil
import lxml.etree as etree

# https://www.ximalaya.com/revision/play/v1/audio?id=246660336&ptype=1

workPath = os.getcwd() + os.sep + '喜马拉雅'

def download(link):
    
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_2) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36',
              'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9'
              }
    for p in range(1, 30):
        albumUrl = link + ('/p%d' % p)
        req = urllib.request.Request(albumUrl, None, headers)
        try:
            with urllib.request.urlopen(req) as response:
                the_page = response.read()
        except urllib.request.URLError as e:
            print(e)
            return None
    
        parser = etree.XMLParser(recover=True)
        root = etree.fromstring(the_page, parser)
        title = root.xpath('//*[@id="award"]/main/div[1]/div[2]/div[1]/div[1]/div[2]/div[2]/h1')[0].text
        
        if p == 1:
            print('Download %s' % title)
            bookpath = workPath + os.sep + title
            if os.path.exists(bookpath):
                if os.path.commonprefix((os.path.realpath(bookpath), workPath)):
                    shutil.rmtree(bookpath)
                else:
                    raise Exception() 
                pass
            os.mkdir(bookpath)
            pass
        audios = root.xpath('//*[@id="anchor_sound_list"]/div[2]/ul/li[*]/div[2]/a')
        if audios == 0 or len(audios) == 0:
            return
        for text in audios:
            audio_title = text.attrib['title']
            print('Download %s' % audio_title)
            audio_path = bookpath + os.sep + audio_title.replace('/', ':') + '.m4a'
            href = text.attrib['href']
            audioId = href[href.rfind('/')+1:]
            audioUrl = 'https://www.ximalaya.com/revision/play/v1/audio?id=%s&ptype=1' % audioId
            req = urllib.request.Request(audioUrl, None, headers)
            try:
                with urllib.request.urlopen(req) as response:
                    the_page = response.read()
            except urllib.request.URLError as e:
                print(e)
                return None

            response = json.loads(the_page)
            urllib.request.urlretrieve (response['data']['src'], audio_path)
            pass
        pass
    return 

    
        
    pass

download('https://www.ximalaya.com/ertong/11106118/')


```

```python
import os
import subprocess

def convertMp3(folder):
    for i in os.listdir(folder):
        if not i.endswith('.m4a'):
            continue
        input_file = folder + os.sep + i
        output_file = folder + os.sep + i[:-4] + '.mp3'
        process = subprocess.Popen('ffmpeg -i "%s" -acodec libmp3lame -aq 2 "%s" -y' % (input_file, output_file), shell=True, stdout=subprocess.PIPE)
        process.wait()
        print('Convert %d %s' % (process.returncode, i))
        if os.path.exists(output_file):
            os.remove(input_file)
            pass
        pass
    pass

convertMp3('喜马拉雅/宝宝巴士·国学儿歌 | 唐诗三百首')
# 非常慢，需要等待
```
